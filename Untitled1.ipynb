```python
def f(t) :
    return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.1)
plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1),'bo',t2,f(t2),'k')
plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

```python
mu = 100
sigma = 15
x = mu + sigma * np.random.randn(10000)

plt.hist(x, 400,density=1, color='g', alpha = 1)
plt.grid(True)
plt.axis([20,200,0,0.03])
plt.text(20.0,0.025, ' meuh=100, $\sigma=15$')
plt.annotate('maximum',xy=(100,0.030),xytext=(140,0.025), arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

```python
n_groups=5
means_men= (20,30,35,27,18)
means_women= (20,32,34,20,25)
index = np.arange(n_groups)
bar_width=0.35
opacity=0.4

fig,ax =plt.subplots()

rects1= ax.bar(index,means_men,bar_width, alpha=opacity,color = 'b', label='Men')
rects2= ax.bar(index+bar_width,means_women,bar_width, alpha=opacity,color = 'r', label='women')

ax.set.xlabel('Groupes')
ax.set.ylabel('Age')
ax.se_sticks(index+0.5+bar_width)
ax.set_xsticklabels(('A','B','C','D','E'))
ax.legend()
ax.show()
```
